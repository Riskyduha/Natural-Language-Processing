```python
import re
import pandas as pd
```

```python
# 1. Kamus Formalisasi (update berkala)
# 'Normalization Lexicon'
slang_dict = {
    "gk": "tidak", "gak": "tidak", "ga": "tidak", "nggak": "tidak",
    "yg": "yang", "utk": "untuk", "dgn": "dengan", "sdh": "sudah",
    "tp": "tapi", "klo": "kalau", "kalo": "kalau", "bgt": "banget",
    "emiten": "perusahaan", "cuan": "untung", "loss": "rugi",
    "skrg": "sekarang", "kmrn": "kemarin", "trus": "terus"
}
```

```python
def normalize_text(text):
    # A. Lowercase
    text = str(text).lower()

    # B. Menangani Elongated Words (Huruf Berulang)
    # Contoh: 'okeeeee' -> 'oke', 'banyaaaaak' -> 'banyak'
    # Regex ini mencari karakter yang sama berturut-turut dan menjadikannya satu karakter
    text = re.sub(r'(.)\1+', r'\1', text)

    # C. Formalisasi Kata Slengekan
    # Pecah jadi kata-kata dulu, lalu cek ke kamus
    words = text.split()
    normalized_words = [slang_dict.get(word, word) for word in words]
    text = " ".join(normalized_words)

    # D. Membersihkan karakter sampah (URL, Mention, Simbol)
    text = re.sub(r'http\s+|www\.\S+|@[A-Za-z0-9_]+|#[A-Za-z0-9_]+', '', text)
    text = re.sub(r'[^a-z\s]', ' ', text) # Hanya sisakan huruf a-z

    # E. Menghapus spasi berlebih
    text = ' '.join(text.split())

    return text
```

```python
# Load dataset
df = pd.read_csv('stock_market.csv')

# Tambahkan kolom baru untuk hasil yang sudah formal
print("Sedang memproses formalisasi teks...")
df['text_formal'] = df['Sentence'].apply(normalize_text)

# Lihat perbandingannya
print("\nHasil Preprocessing (5 data pertama):")
display(df[['Sentence', 'text_formal']].head())
```
